```python
# Playlist fetcher for checking what playlists a song-link I want to add is within

import os
from pprint import pprint

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=os.getenv('SPOTIPY_CLIENT_ID_CONTAINED_PLAYLISTS_FETCHER'), client_secret=os.getenv('SPOTIPY_CLIENT_SECRET_CONTAINED_PLAYLISTS_FETCHER'))
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

library_account = os.getenv('SPOTIFY_PLUS_SECONDARY_ACCOUNT_USER_ID')
```

```python
simplified_playlists = []

simplified_immediate_to_sort = {}
simplified_library_to_sort = {}

simplified_genres = []
simplified_archived_mixtapes = []
simplified_archived_records = []

current_page = sp.user_playlists(library_account)
while current_page:
    for playlist in current_page['items']:
        new_simplified = {}
        new_simplified['id'] = playlist['id']
        new_simplified['name'] = playlist['name']

        simplified_playlists.append(new_simplified)

        if new_simplified['name'].startswith('[G]'):
            simplified_genres.append(new_simplified)
        elif new_simplified['name'].startswith('[AM]'):
            simplified_archived_mixtapes.append(new_simplified)
        elif new_simplified['name'].startswith('[AR]'):
            simplified_archived_records.append(new_simplified)
        elif new_simplified['name'].startswith('[1]'):
            simplified_immediate_to_sort = new_simplified
        elif new_simplified['name'].startswith('[2]'):
            simplified_library_to_sort = new_simplified

    if current_page['next']:
        current_page = sp.next(current_page)
    else:
        current_page = None
```

```python


current_page = sp.playlist_items(f'spotify:playlist:{simplified_library_to_sort["id"]}', fields='items.track.id,items.track.name,next')
while current_page:
    if current_page['next']:
        current_page = sp.next(current_page)
    else:
        current_page = None
```
